## Create a "total" column

```py
df['Total'] = df.iloc[:, 4:10].sum(axis=1)
```

This code creates a new column called `Total` in the pandas DataFrame `df`, and assigns it the sum of values in each row for columns 4 to 9 (i.e., column indices 4, 5, 6, 7, 8, and 9).

*Like a spreadsheet.*

The **`.iloc`** indexer is used to select a subset of the DataFrame based on its integer location.

In this case, **`df.iloc[:, 4:10]`** selects all rows (`:`) and columns from index 4 up to, but not including, index 10 (`4:10`).

This subset of the DataFrame is a DataFrame itself, which contains only the columns of interest for the calculation of the total.

The `sum()` function is applied to this DataFrame along the rows (`axis=1`), which computes the sum of values in each row, resulting in a Series of **row-wise sums.**

Finally, this Series is assigned to the **`Total`** column of the original DataFrame, which adds a new column to the DataFrame containing the row-wise sums calculated from the selected columns.

# Filters

The `&` operator combines multiple conditions into a single boolean expression that is true only when all conditions are true.

So, the code is selecting rows from `df` where the following conditions are all true:

* The 'Type 1' column is equal to 'Grass'.
* The 'Type 2' column is equal to 'Poison'.
* The 'HP' column is greater than 70.

In other words, the code is selecting all the Grass/Poison Pokemon that have an HP greater than 70.

````python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
````

The boolean expression contains three parts connected by the logical operator "&":

1. `df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)` - this part selects all the rows where the 'Type 1' column contains the word 'fire' or 'grass', regardless of the case (the `flags=re.I` argument specifies that the search should be case-insensitive), and using regular expressions (`regex=True` argument).


2. The `|` character in the regular expression specifies an "or" operation, which means that the expression matches if either "fire" or "grass" appears in the 'Type 1' column.


3. Finally, `df.loc[]` selects only the rows where the boolean expression is True, and returns a new DataFrame containing only those rows.


So, this code is selecting all the rows from the df DataFrame where the 'Type 1' column contains either the word 'fire' or 'grass', case insensitive.

# Data Aggregation

```py
df.groupby(['Type 1']) \
    .mean() \
    .sort_values(
    'Defense',
    ascending=False
)
```

This code is performing a data aggregation operation on a pandas DataFrame called df. Specifically, it is **grouping the rows** of the DataFrame by the unique values in the **'Type 1'** column, and then computing the **mean of each numerical column** for each group. Finally, it **sorts** the resulting DataFrame by the 'Defense' column in descending order.

Here's the breakdown:

* `df.groupby(['Type 1'])`: This groups the rows of the DataFrame by the unique values in the 'Type 1' column, creating a pandas GroupBy object. This object contains the groups of rows that share the same value in the 'Type 1' column.


* `.mean()`: This computes the mean of each numerical column for each group of rows. This creates a new DataFrame where each row corresponds to a unique value in the 'Type 1' column, and the columns are the mean values for each numerical column.


* `.sort_values('Defense', ascending=False)`: This sorts the resulting DataFrame by the 'Defense' column in descending order (from highest to lowest value). The resulting DataFrame is returned with the rows sorted by the mean Defense value for each Type 1 group.


Get all the distinct types, averate each numerical column, then sort by defense, desc.

This can be used to get insights into the defensive capabilities of different Pokemon types.

SQL:

```sql
SELECT   'Type 1',
         AVG('HP')      AS 'mean_HP',
         AVG('Attack')  AS 'mean_Attack',
         AVG('Defense') AS 'mean_Defense'
FROM     'table_name'
GROUP BY 'Type 1'
ORDER BY 'mean_Defense' DESC 
```


````python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
````

## Reordering DataFrame

This code reorders the columns.

```py
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

The new order of the columns is:

* The first four columns of the original DataFrame (`cols[0:4]`)
* The last column of the original DataFrame (`[cols[-1]]`)
* The columns in the original DataFrame starting from index 4 up to but not including index 12 (`cols[4:12]`)

The `+` operator concatenates the three lists of column names together into a single list, which is then used to index the original DataFrame. The result is a new DataFrame with the columns in the new order.
